## Quick Sort

```python
import numpy as np
import random 
```

```python
def quick_sort(arr, left, right):
    if left >= right:
        return

    i, j = left, right
    pivot = arr[left]

    # 挖坑法 partition
    while i < j:
        while i < j and arr[j] >= pivot:
            j -= 1
        if i < j:
            arr[i] = arr[j]
            i += 1

        while i < j and arr[i] <= pivot:
            i += 1
        if i < j:
            arr[j] = arr[i]
            j -= 1

    arr[i] = pivot
    pivot_idx = i

    quick_sort(arr, left, pivot_idx - 1)
    quick_sort(arr, pivot_idx + 1, right)

    return arr


random.seed(42)
arr=[random.randint(0,100) for _ in range(10)]
print("Original array:", arr)
left=0
right=len(arr)-1
sorted_arr=quick_sort(arr,left,right)
print("Sorted array:", sorted_arr)
```
